# 1. Basic Plotting

```python
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')


# Let's take a data representing monthly website traffic (in thousands)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
traffic_A = [150, 200, 180, 220, 250, 160]

# Create a line plot with custom appearance
plt.plot(months, traffic_A, marker='o', linestyle='--', color='g')

# Add labels and a title
plt.xlabel('Month')
plt.ylabel('Monthly Traffic (in Thousands)')
plt.title('Monthly Website Traffic')

# Add grid to the plot
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Traffic in City B
traffic_B = [150, 190, 200, 215, 350, 180]

# plt.xkcd() # Trick 1

# Create a line plot for Traffic A with a blue line and circular markers
plt.plot(months, traffic_A, marker='o', linestyle='-', color='blue', label='Traffic A')

# Create a line plot for Traffic B with a red dashed line and square markers
plt.plot(months, traffic_B, marker='s', linestyle='--', color='red', label='Traffic B')

# plt.fill_between(months, traffic_A, traffic_B, alpha=0.3, color='purple') # Trick 2

# Add labels and a title
plt.xlabel('Month')
plt.ylabel('Monthly Traffic (in Thousands)')
plt.title('Monthly Traffic Comparison')


# Display a legend to differentiate between Product A and Product B
plt.legend()

# Display the plot
plt.show()

# plt.style.use('default') # Trick 1
```

# 2. Bar Chart

Bar charts represent categorical data with rectangular bars, where the length or height of each bar represents a value.

Few Use Cases:
1. Comparing sales performance of different products.
2. Showing population distribution by country.

Eg: Multi Bar plot in a single Graph

```python
# Traffic in City C
traffic_C = [250, 150, 120, 125, 430, 190]

# Create an array for the x-axis positions
x = np.arange(len(months))

# Width of the bars, we need it as we intend to plot multiple bars.
bar_width = 0.2

# Create bars for City A,
# such that it will be placed to the left.
plt.bar(x - bar_width, traffic_A, width=bar_width, label='City A', color='skyblue')

# Create bars for City B
plt.bar(x, traffic_B, width=bar_width, label='City B', color='lightcoral')

# Create bars for City C
# such that it will be placed to the right.
plt.bar(x + bar_width, traffic_C, width=bar_width, label='City C', color='lightgreen')

# Alternative
#plt.barh(x - bar_width, traffic_A, height=bar_width, label='City A', color='skyblue')
#plt.barh(x, traffic_B, height=bar_width, label='City B', color='lightcoral')
#plt.barh(x + bar_width, traffic_C, height=bar_width, label='City C', color='lightgreen')

# Add labels, a title, and a legend
plt.xlabel('Traffic Months')
plt.ylabel('Monthly Traffic (in Thousands)')
plt.title('Monthly Traffic Comparison')

# To show the category names at x-axis positions.
plt.xticks(x, months)
plt.legend()

# Display the plot
plt.show()
```

In two subplots...

```python
# Create a 2x1 grid of subplots
fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(10, 8))

# Plot the first subplot (top-left) - axes[0,0]
axes[0].bar(x - bar_width, traffic_A, width=bar_width, label='City A', color='skyblue')
axes[0].bar(x, traffic_B, width=bar_width, label='City B', color='lightcoral')
axes[0].bar(x + bar_width, traffic_C, width=bar_width, label='City C', color='lightgreen')
axes[0].set_title('Monthly Traffic Comparison')
axes[0].set_xticks(x, months)
axes[0].legend()



# Plot the second subplot (top-right) - axes[0,1]
axes[1].barh(x - bar_width, traffic_A, height=bar_width, label='City A', color='skyblue')
axes[1].barh(x, traffic_B, height=bar_width, label='City B', color='lightcoral')
axes[1].barh(x + bar_width, traffic_C, height=bar_width, label='City C', color='lightgreen')
axes[1].set_title('Monthly Traffic Comparison')
axes[1].set_yticks(x, months)
axes[1].legend()


# Adjust spacing between subplots
plt.tight_layout()


# Display the subplots
plt.show()
```

# 3. Histograms

Histograms are used to visualize the distribution of continuous or numerical data and they help us identify patterns in data. In a histogram plot the data is grouped into bins, and the height of each bar represents the frequency or count of data points in that bin. It takes the lower and upper limits of the given data and divides it into the no of bins given.

Few Use Cases:
1. Analyzing age distribution in a population.
2. Examining exam score distribution in a classroom.

```python
# Total traffic per Months
traffic_total = [sum(x) for x in zip(traffic_A,traffic_B,traffic_C)]

lower_bound = int(min(traffic_total)*0.9)
upper_bound = int(max(traffic_total)*1.1)
ranging = int((upper_bound - lower_bound)/2)

# Custom bin ranges
bin_ranges = range(lower_bound, upper_bound, ranging)

# Create a histogram with custom bin ranges by assigning it to bins.
plt.hist(traffic_total, bins=bin_ranges, color='lightblue', edgecolor='black', alpha=0.7)

# Add labels and a title
plt.xlabel('Total Traffic per Months')
plt.ylabel('Frequency (Months)')
plt.title('Total Traffic per Months Histogram with Custom Bins')

# Calculate and add a median line
median_score = np.median(traffic_total)
plt.axvline(median_score, color='red', linestyle='dashed', linewidth=2, label=f'Median Score: {median_score}')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

# 4. Scatter plots

Scatter plots display individual data points as dots on a two-dimensional plane. They are used to explore relationships or correlations between two numerical variables.

Few Use Cases:
1. Investigating the relationship between study hours and exam scores.
2. Analyzing the correlation between temperature and ice cream sales.

```python
# Scaling months for point sizes in the scatter plot
point_sizes = [size * 5 for size in traffic_total]

# To Create a scatter plot with different point sizes,
# Here alpha controls the intensity of the color.
plt.scatter(traffic_A, traffic_C, s=point_sizes, c='skyblue', alpha=0.7, edgecolors='b')

# Add labels, a title, and a legend
plt.xlabel('Traffic City A')
plt.ylabel('Traffic City B')
plt.title('Relationship between Traffic Cities with respect total Traffic')

# To Display the plot
plt.show()
```

# 5. Pie Charts

Pie charts represent parts of a whole as slices of a circular pie. They are suitable for showing the composition of a single categorical variable in terms of percentages. But this wont look good when there are more than six categories as they get clumsy, in such cases horizontal bar might be preferred.

Few Use Cases:
1. Displaying the distribution of a budget by expense categories.
2. Showing the market share of various smartphone brands.

Eg: Exploding a particular segment for better storytelling during presentations.

```python
# Explode a specific segment (e.g., 'April'),
# second value (0.1) is the amount by which the segment 'April'.
explode = (0, 0, 0, 0.1, 0, 0)

# Create a pie chart with explode and shadow
plt.pie(traffic_total, labels=months, explode=explode, shadow=True, autopct='%1.1f%%')
plt.title('Total Traffic per Month')

# Display the plot
plt.show()
```

# 6. Box Plot

Box plots summarize the distribution of numerical data by displaying quartiles, outliers, and potential skewness. They provide insights into data spread, central tendency, and variability. Box plots are especially useful for identifying outliers and comparing distributions.

Few Use Cases:
1. Analyzing the distribution of salaries in a company.
2. Assessing the variability of housing prices in different neighborhoods.



```python
# Create a box plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.boxplot(traffic_total, vert=False, patch_artist=True,
            boxprops={'facecolor': 'lightblue'},
            flierprops={'marker': 'o', 'markerfacecolor': 'red',
                        'markeredgecolor': 'red'})

# Add labels and a title
plt.xlabel('Total Traffic')
plt.title('Box Plot with Outliers')

# Display the plot
plt.grid(True)  # Add a grid for better readability
plt.show()
```

# 7. Heatmap

Heatmap is a visualization for correlation matrix, which will give us a sense of how each variable is correlated with the other variable.

```python
correlation_matrix = np.corrcoef([traffic_A, traffic_B, traffic_C, traffic_total])

# Create a heatmap for the correlation matrix
plt.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1, aspect='auto', origin='upper')

# Add a colorbar with the following commands
cbar = plt.colorbar()
cbar.set_label('Correlation', rotation=270, labelpad=20)

# Add labels and a title
plt.title('Correlation Matrix Heatmap')
plt.xticks(range(len(correlation_matrix)), ['A','B','C','TOT'])
plt.yticks(range(len(correlation_matrix)), ['A','B','C','TOT'])

plt.show()
```

# 8. Stack Plot



```python
# Create a stack plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.stackplot(months, traffic_A, traffic_B, traffic_C, labels=['City A', 'City B', 'City C'],
              colors=['blue', 'green', 'red'], alpha=0.7)

# Add labels, legend, and title
plt.xlabel('Months')
plt.ylabel('Traffic (Thousands)')
plt.title('Total Traffic by Months')
plt.legend(loc='upper left')

# Display the plot
plt.grid(True)
plt.show()
```
